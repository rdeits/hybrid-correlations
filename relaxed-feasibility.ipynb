```julia
using Colors
using AxisArrays
cmap = colormap("RdBu", 101)
function colorize(value)::RGB{Float64}
    isnan(value) ? colorant"white" : cmap[round(Int, value * 50 + 51)]
end
```

```julia
using Compat.Iterators: product
```

```julia
double_integrator = c.MPC.discretize(c.MPC.CTLinearSytstem([0. 1; 0 0], [0. 1]'), 0.1)
samples = c.collect_data(double_integrator, 500);
correlations = c.correlate(samples);
```
