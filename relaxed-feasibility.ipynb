```julia
using Colors
using AxisArrays
cmap = colormap("RdBu", 101)
function colorize(value)::RGB{Float64}
    isnan(value) ? colorant"white" : cmap[round(Int, value * 50 + 51)]
end
```

```julia
include("main.jl")
```

```julia
double_integrator = c.MPC.discretize(c.MPC.CTLinearSytstem([0. 1; 0 0], [0. 1]'), 0.1)
samples = c.collect_data(double_integrator, 1000);
correlations = c.correlate(samples);
```

```julia
vcat(hcat(colorize.(correlations[
    Axis{:constraint}(:limb_ub),
    Axis{:contact_side}(:right),
    Axis{:polarity}(:on)
        ]), 
fill(RGB{Float64}(0, 0, 0), 9, 1),
colorize.(correlations[
    Axis{:constraint}(:limb_ub),
    Axis{:contact_side}(:right),
    Axis{:polarity}(:off)
        ])),
fill(RGB{Float64}(0, 0, 0), 1, 21),
hcat(colorize.(correlations[
    Axis{:constraint}(:limb_ub),
    Axis{:contact_side}(:left),
    Axis{:polarity}(:on)
        ]), 
fill(RGB{Float64}(0, 0, 0), 9, 1),
colorize.(correlations[
    Axis{:constraint}(:limb_ub),
    Axis{:contact_side}(:left),
    Axis{:polarity}(:off)
        ])))
```
